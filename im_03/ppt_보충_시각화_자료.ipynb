```python
import matplotlib.pyplot as plt
import numpy as np

# 가정 데이터
years = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
total_revenue = [15, 18, 22, 27, 33, 40, 48]  # 단위: 억
loan_revenue = [4, 5, 7, 10, 13, 17, 22]  # 단위: 억

# x 위치 조정
x = np.arange(len(years))
width = 0.35

# 막대그래프 생성
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, total_revenue, width, label='총 수익성')
plt.bar(x + width/2, loan_revenue, width, label='개인 대출 수익성')

# 그래프 꾸미기
plt.title('회사의 연도별 총 수익성과 개인 대출 수익성')
plt.xlabel('연도')
plt.ylabel('수익성 (억)')
plt.xticks(x, years)
plt.legend()

plt.savefig('연도별_개인_대출_수익성_막대그래프.png')
# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

# 가정 데이터
quarters = ['1분기', '2분기', '3분기', '4분기']
delinquency_rates = [7.6, 8.4, 8.9, 9.2]  # 단위: %

# 그래프 생성
plt.figure(figsize=(10, 6))
plt.plot(quarters, delinquency_rates, marker='o', label='연체율',color='red')

# 그래프 꾸미기
plt.title('대출금액에 대한 분기별 연체율')
plt.xlabel('분기')
plt.ylabel('당사은행 연체율 (%)')
plt.legend()
plt.grid(False)
plt.savefig('자사_분기별_연체율_선그래프.png')
# 그래프 표시
plt.show()
```
