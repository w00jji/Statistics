```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3'],})
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7'],})
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11'],})
```

```python
rest =pd.concat([df1,df2,df3])
rest
```

```python
# concat의 parameter 변경
rest =pd.concat([df1,df2,df3],axis=1)
rest
```

```python
# ignore_index = True : 기존 index가 아닌 새로운 index로 재생성
# axis=0 세로 , axis=1 가로
rest =pd.concat([df1,df2,df3],axis=0,ignore_index=True)
rest
```

```python
# concat 결과 데이터의 변수명 생성
rest.rename(columns={'0':'A',
                    '1':'B',
                    '2':'C',
                    '3':'D',
                    '4':'E',
                    '5':'F',
                    '6':'G',
'7':'H','8':'I','9':'J','10':'K','11':'L',},inplace=True)
rest
```
